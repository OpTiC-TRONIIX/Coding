### PageRank

What is the most important website on the internet? Who is the "key player" on a sports team? Which countries are the most central players in the world economy? There is no one correct answer to any of these questions, but there is a most profitable one. [PageRank](https://en.wikipedia.org/wiki/PageRank) is an algorithm for ranking individual elements of complex systems, invited by Sergey Brin and Larry Page. It was the first and most famous algorithm used by the Google Search engine, and it is fair to say that the internet as we know it today would not exist without PageRank. 

In this assignment, we will implement PageRank. There are many good ways to implement this algorithm, but in this assignment we will use our newfound skills with object-oriented programming and iterators. 

### How it works

For the purposes of this example, let's assume that we are talking about webpages. PageRank works by allowing a "random surfer" to move around webpages by following links. Each time the surfer lands on a page, it then looks for all the links on that page. It then picks one at random and follows it, thereby arriving at the next page, where the process repeats. Eventually, the surfer will visit all the pages one or more times. Pages that the surfer visits more frequently have higher *PageRank scores*. Because the surfer moves between linked pages, PageRank expresses an intuitive idea: **important pages are linked to other important pages.** [This diagram](https://en.wikipedia.org/wiki/PageRank#/media/File:PageRanks-Example.jpg) from Wikipedia gives a nice illustration. Note that more important webpages (higher PageRank) tend to be connected to other important webpages.

#### Your Solution

For problem (D), I wasn't sure how to start. I had to plan this question step-by-step, and once done, I began working on the code. For the function, I needed to return a dictionary, so, of course, an empty dictionary needed to be created. Further, since unique elements were being used as keys in the dictionary, I decided to use the special property of sets to extract these unique elements. As for the values of the dictionary, I struggled to figure out how to extract all elements associated with a given 0th position element in the tuple. I knew a for-loop and if statement were necessary but implementing them was convoluted. I realized that an additional iterating variable was necessary, but it should only iterate once all values in the list of data have been compared to the tuples. That is, every 0th position element in the list of data was compared to the 0th position element of the unique-elements list. If they match, append the 1st position element of the list of data to a list. If they don't, the extra iterating variable is increased by one and the list becomes empty. Again, all 0th position elements in the list of data are compared to the 1th position element of the unique-elements list. This system works since the lists were all sorted beforehand. To finish, I simply returned the dictionary and tested whether it works. It does.

````python
#Problem D
def data_to_dictionary(data):
    """
    Takes in a parameter data, which is a list of tuples.
    A set and a list are formed from the inputted data.
    Converts the data from the set into another list to extract unique tuples.
    The resulting lists are then sorted (for ease of use).
    Another list is created from the list of unique tuples containing the 0th element in each tuple.
    Again, this list of tuple elements is converted to a set and back to a list to extract unique elements, and then sorted.
    Each 1st position element associated with every 0th position element is extracted.
    A dictionary is created accordingly.
    """
    D = {}
    dataset = set(data) #Converting data to set to identify unique tuples
    datalist = list(data)
    datalist.sort()
    datalistspecial = list(dataset) #Converting set to list of unique tuples
    datalistspecial.sort()
    j = 0 #Counter for creating the dictionary
    L = [] #List for elements in second position in each tuple in datalist
    L2 = [] #List for elements corresponding to first position in each tuple in datalist
    
    #For loop to generate L2 defined above.
    for i in range(len(datalistspecial)):
        L2.append(datalistspecial[i][0])
    L2set = set(L2)
    L3 = list(L2set) #List of unique elements in first position of tuple in datalist
    L3.sort()
    #Iterating through the list of data
    for i in range(len(datalist)):
        #If the 0th element matches the jth element of L3, append it to L
        if datalist[i][0] == L3[j]:
            L.append(datalist[i][1])
        #Else, increase j by 1, make the list empty, and append the 1st position element of datalist to L.
        else:
            j += 1
            L = []
            L.append(datalist[i][1])
        #Create the dictionary by updating it.
        D.update({datalist[i][0] : L})
    #Returning the dictionary
    return D

#Testing the function with sample output from above
dat = [("a", "b"), 
        ("a", "b"), 
        ("a", "c"),
        ("b", "c"),
        ("b", "a")]
print(data_to_dictionary(dat))
````

````python
#Problem E
class PR_DiGraph:
    """
    A custom class that creates sets of data from the output of the data_to_dictionary funtion above.
    These sets of data are used to generate a PageRank for the original data used.
    """
    def __init__(self, data, iteration_limit):
        """
        Constructs instance variables
        """
        self.link_dict = data_to_dictionary(data) #Initializing the link_dict
        self.iteration_limit = iteration_limit #Initializing the iteration_limit
        
    def linked_by(self, x):
        """
        A method that returns the values of the output of data_to_dictionary (applied to data) for some value, x.
        """
        return self.link_dict[x] #Returning the values for some key, x
    
    def __iter__(self):
        """
        Returns a PR_Iterator object defined in the next problem.
        """
        return PR_Iterator(self)

#Testing class
Dtest = PR_DiGraph(data, iteration_limit = 10000)
Dtest.linked_by("SIN")
````

#### Your Solution

For part (F) below, I immediately noticed that the code in the class followed almost directly from the lecture notes and Homework 1. Initializing the instance variable, again, was fairly easy. However, I made a mistake and initialized the current_state instead of a DiGraph Object, which was later rectified. The follow_link function was pretty easy to 'draft' but issues arose with repetitive output; this turned out to be a fault with placement of code outside of the try-except block. As a result, the necessary code was put inside the try-except block and the issue was corrected. The teleport function was initially created directly from the PR_Digraph Object, but it was later made a little more explicit (step-by-step) with a list for ease of understanding. The next method followed from fundamental for-loop logic, the lecture notes and the code given above. The current_state was returned at the end of the next method, and the output, after much testing, came out alright.

````python
#Problem F
import random

class PR_Iterator:
    """
    An iterator class that enables iteration using a PR_Digraph Object.
    """
    def __init__(self, PRD):
        """
        Constructs and initialized instance variables.
        """
        self.i = 0 #Initializing i with 0
        self.current_state = "SIN" #Initializing current_state to "SIN"
        self.PRD = PRD #Constructing a PR_DiGraph Object
    
    def teleport(self):
        """
        Creates list of keys from the PR_DiGraph object's dictionary.
        The current state is set to some random key from this list.
        """
        L = [] #Empty list
        for keys in self.PRD.link_dict.keys():
            L.append(keys) #Appending keys to list
        #The current_state becomes some random key from L
        self.current_state = random.choice(L)
    
    def follow_link(self):
        """
        Picks a random values associated with a given key.
        'Toggles' between states based on random choice from keys, values in the PR_DiGraph Object.
        Overcomes potential KeyError(s) using try-except blocks.
        """
        #Try the following code
        try:
            self.next_state = random.choice(self.PRD.linked_by(self.current_state))
            if self.next_state not in (self.PRD.link_dict.keys()):
                self.teleport()
            if self.next_state != self.current_state:
                self.current_state = self.next_state
            elif self.next_state == self.current_state:
                self.teleport()
        #If a KeyError arises from random value not being in data or repeat values, teleport and form a new current_state
        except KeyError:
            self.teleport()
    
    def __next__(self):
        """
        Iterates through i and raises a StopIteration when i equals the user-inputted iteration limit.
        Uses the follow_link function 85% of the time and teleport function 15% of the time.
        Increments i.
        Returns the current_state.
        """
        #Stop when i equals the iteration limit
        if self.i >= self.PRD.iteration_limit:
            raise StopIteration
        #Run follow_link 85% of the time
        if random.random() < 0.85:
            self.follow_link()
        #Run teleport 15% of the time
        else:
            self.teleport()
        #Increment i by 1
        self.i += 1
        
        #Return the current_state
        return self.current_state
````
